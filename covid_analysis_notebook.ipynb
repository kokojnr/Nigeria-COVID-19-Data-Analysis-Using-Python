NCDC Website scrap
Website - https://covid19.ncdc.gov.ng/ experiencing downtime
Data Obtained from wikipedia : https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Nigeria


```python
# Send a GET request to the Wikipedia page
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Nigeria"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# Find the section heading element containing the specific text
target_text = "Confirmed COVID-19 cases in Nigeria by state"
target_element = soup.find("span", {"id": "By_state"}).parent

# Find the table within the section
table = target_element.find_next_sibling("table", class_="wikitable")

# Extract the table rows
rows = table.find_all("tr")

# Extract and print the table headers
headers = [header.text.strip() for header in rows[0].find_all("th")]

# Extract and print the table data
data = []
states = []
for row in rows[1:]:
    cells = row.find_all(["td", "th"])
    state_name = cells[0].text.strip()
    states.append(state_name)
    row_data = [cell.text.strip() for cell in cells[1:]]
    # Handling missing data by filling empty cells with NaN
    while len(row_data) < len(headers) - 1:
        row_data.append(pd.NA)
    data.append(row_data)

# Create a DataFrame
df = pd.DataFrame(data, columns=headers[1:])  # Exclude the first header for the State column

# Add the 'States' column to the DataFrame
df.insert(0, "State", states)



```

```python
# Calculate daily infection rate
df = nig_df.copy()        #github data
df = df.sort_values(by='Date')
df['DailyInfectionRate'] = df['Confirmed'].diff()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['DailyInfectionRate'], label='Daily Infection Rate', marker='o', color='orange')

plt.xlabel('Date')
plt.ylabel('Daily Infection Rate')
plt.title('Daily Infection Rate Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Find the date with the maximum infection rate
max_infection_date = df.loc[df['DailyInfectionRate'].idxmax(), 'Date']
max_infection_rate = df['DailyInfectionRate'].max()

# Print the result
print(f"Maximum Infection Rate: {max_infection_rate} on {max_infection_date}")
```

```python
# Define date ranges
ranges = [
    ('2020-01-01', '2020-03-29'), # pre lockdown
    ('2020-03-30', '2020-06-1'),  # lockdown
    ('2020-06-02', '2020-06-29'), # ease phase one
    ('2020-06-30', '2020-07-27'), # ease phase two
]

# Convert date ranges to datetime objects
date_ranges = [(pd.to_datetime(start), pd.to_datetime(end)) for start, end in ranges]


# Create bins and labels
bins = [pd.to_datetime(start) for start, _ in date_ranges] + [pd.to_datetime('2020-07-28')]  # Add an upper bound for the last bin
labels = ['Pre Lockdown', 'Lockdown', 'Phase 1', 'Phase 2']

# Cut the data into bins and assign labels
nig_df['Phase'] = pd.cut(nig_df['Date'], bins=bins, labels=labels, right=False)

# Calculate total confirmed cases for each phase
plot_df = nig_df.groupby('Phase')['Confirmed'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(plot_df['Phase'], plot_df['Confirmed'], color='skyblue')
plt.title('Total Confirmed Cases for Different Phases')
plt.xlabel('Lockdown Phases')
plt.ylabel('Total Confirmed Cases')
plt.show()
```
